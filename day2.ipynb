```python
import pandas as pd

# Direct URL for Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Show first 5 rows
print(df.head())

#basic structure
#see how big the dataset is 
print("Rows,columns:",df.shape)
# info abt each column
df.info()
```

```python
#quick stats fr numbers(age,fare,etc.)
df.describe()
```

```python
#check missing values column by column
df.isnull().sum()
```

```python
#see only female passengers
df[df['Sex']=='female'].head()
```

```python
#sort passengers by age (youngest first)
df.sort_values('Age').head()
```

```python
#show only some columns
df[['Name','Age','Fare']].head()
```

```python
#lets visualize it into bar grpahs
import seaborn as sns
import matplotlib.pyplot as plt


#gender distribution
sns.countplot(x='Sex', data=df)
plt.show()

#survival by class(who survived in each class)
sns.countplot(x='Pclass',hue='Survived', data=df)
plt.show()
```
